```python
# Import necessary libraries
import matplotlib.pyplot as plt
import pandas as pd
import json
from IPython.display import display

test_results = "../../models/Hybrid_Concat/Hybrid_Concat_Freeze/test_results.json"
training_metrics = "../../models/Hybrid_Concat/Hybrid_Concat_Freeze/training_history.json"
# Uncomment these lines to read from actual files
with open(test_results, 'r') as f:
    test_results = json.load(f)
with open(training_metrics, 'r') as f:
    training_metrics = json.load(f)

# Create table for test results
test_df = pd.DataFrame([test_results], index=['Results'])
print("Test Results Table:")
display(test_df)

# Set up figure for plots
epochs = range(1, len(training_metrics['train_loss']) + 1)

# Plot 1: Loss
plt.figure(figsize=(10, 6))
plt.plot(epochs, training_metrics['train_loss'], 'b-o', label='Training Loss')
plt.plot(epochs, training_metrics['val_loss'], 'r-o', label='Validation Loss')
plt.title('Training and Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)
plt.show()

# Plot 2: Intent F1
plt.figure(figsize=(10, 6))
plt.plot(epochs, training_metrics['train_intent_f1'], 'b-o', label='Training Intent F1')
plt.plot(epochs, training_metrics['val_intent_f1'], 'r-o', label='Validation Intent F1')
plt.title('Intent F1 Score')
plt.xlabel('Epochs')
plt.ylabel('F1 Score')
plt.legend()
plt.grid(True)
plt.show()

# Plot 3: Category F1
plt.figure(figsize=(10, 6))
plt.plot(epochs, training_metrics['train_category_f1'], 'b-o', label='Training Category F1')
plt.plot(epochs, training_metrics['val_category_f1'], 'r-o', label='Validation Category F1')
plt.title('Category F1 Score')
plt.xlabel('Epochs')
plt.ylabel('F1 Score')
plt.legend()
plt.grid(True)
plt.show()

# Plot 4: NER F1
plt.figure(figsize=(10, 6))
plt.plot(epochs, training_metrics['train_ner_f1'], 'b-o', label='Training NER F1')
plt.plot(epochs, training_metrics['val_ner_f1'], 'r-o', label='Validation NER F1')
plt.title('NER F1 Score')
plt.xlabel('Epochs')
plt.ylabel('F1 Score')
plt.legend()
plt.grid(True)
plt.show()
```
